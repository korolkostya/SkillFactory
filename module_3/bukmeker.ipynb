```python
def time_day(x):
    x=int(x)
    if (x >= 0 and x < 6):
        return 'ночь'
    elif (x >= 6 and x < 12):
        return 'утро'
    elif (x >= 12 and x < 18):
        return 'день'
    elif (x >= 18 and x < 24):
        return 'вечер'
```

```python
def fillna_win(row):  
    if pd.isna(row.win):
        if row['bet'] == 0:
            return 0
        else:
            return row['bet'] * -1
    else:
        return row['win']
    
  
# Применяем функцию  
new_win = log5.apply(lambda row: fillna_win(row), axis=1)  
# Заменяем старый столбец с пропусками на новый без пропусков  
log5['win'] = new_win
log5
```

```python
#11.9.3
log = pd.read_csv('log.csv', header = None)
log.columns = ['user_id', 'time', 'bet', 'win']
log = log[log.user_id != '#error']
log.time = log.time.str.replace('[', '')
log.user_id = log.user_id.str.split(' - ').apply(lambda x: x[1])
log['time'] = pd.to_datetime(log['time'])
log['bet'] = log['bet'].fillna(0)
df2 = log[['user_id', 'bet', 'time']]
df3 = df2[(df2.bet == 0)]  # df3 включает ВСЕХ кто появился в базе
df3 = df3.groupby('user_id').aggregate({'time': 'min'})  # дата 1-го появления
df4 = df2[(df2.bet > 0)]  # df4 включат только тех кто СДЕЛАЛ СТАВКУ
df4 = df4.groupby('user_id').aggregate({'time': 'min'})  # дата 1-й ставки
df5 = pd.merge(df3, df4, on='user_id')
df5['my_period'] = df5['time_y'] - df5['time_x']  # создаем столбец = искомый период
df6 = df5[(df5['my_period'] > pd.Timedelta(0))] # убираем ОТРИЦАТЕЛЬНЫЕ периоды
df6['my_period'].mean()   # возвращает  49 дней
# а если НЕ убирать отрицательные периоды (где нулевая ставка была ПОСЛЕ ненулевой)
df5['my_period'].mean()   # тогда возвращает   "правильный"   ответ  = 46 дней
```
