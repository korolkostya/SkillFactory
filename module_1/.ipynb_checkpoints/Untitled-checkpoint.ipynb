```python
log = pd.read_csv('log.xls')
log.columns = ['user_id', 'time', 'bet', 'win']
log = log[log['user_id'] != '#error']
def users_cleaner(user):
    if "Запись пользователя № - user_" in user:
        return user.split()[-1]
log['user_id'] = log.user_id.apply(users_cleaner)
log.time = log.time.str.replace('[', '')
log
```
